```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('tu_list.csv', index_col=0)
df2 = pd.read_csv('tu_list2.csv', index_col=0)
```

```python
df = pd.concat([df1, df2], sort=False)
```

```python
# A = np.unique(df.fillna('').iloc[:,0])
A = np.unique(df['มหาวิทยาลัย'].fillna(''))
A
```

```python
from fixwords import fixranks, fixuni
#df.iloc[:,0] = df.fillna('').iloc[:,0].apply(fixranks)
df['มหาวิทยาลัย'] = df['มหาวิทยาลัย'].fillna('').apply(fixuni)
```
