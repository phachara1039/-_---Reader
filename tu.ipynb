```python
import pandas as pd
url = ("https://github.com/phachara1039/-_---Reader/raw/master/%E0%B8%A3%E0%B8%B2%E0%B8%A2%E0%B8%8A%E0%B8%B7%E0%B9%88%E0%B8%AD%E0%B8%AD%E0%B8%B2%E0%B8%88%E0%B8%B2%E0%B8%A3%E0%B8%A2%E0%B9%8C%E0%B8%A1%E0%B8%AB%E0%B8%B2%E0%B8%A7%E0%B8%B4%E0%B8%97%E0%B8%A2%E0%B8%B2%E0%B8%A5%E0%B8%B1%E0%B8%A2_%E0%B9%80%E0%B8%A7%E0%B9%87%E0%B8%9A%E0%B9%84%E0%B8%8B%E0%B8%95%E0%B9%8C.xlsx")
xl = pd.ExcelFile(url)
```

```python
cols = ['ตำแหน่งทางวิชาการ', 'ตำแหน่งจากการจบการศึกษาระดับปริญญาเอก', 'ชื่อ', 'สกุล', 'คุณวุฒิ', 'มหาวิทยาลัย', 'คณะ', 'คามเชี่ยวชาญ', \
              'สถานที่ติดต่อ', 'ช่องทางการติดต่อ (เบอร์โทรศัพท์)', 'ช่องทางการติดต่อ (e-mail address)', 'สถานะ']
new_cols = ['ตำแหน่ง', 'ตำแหน่ง ป.เอก', 'ชื่อ', 'นามสกุล', 'คุณวุฒิ', 'มหาวิทยาลัย', 'สาขาวิชา', 'คามเชี่ยวชาญ', \
              'สถานที่ติดต่อ', 'โทร', 'email', 'สถานะ']
```

```python
dfs = None
for i, sheet in enumerate(xl.sheet_names):
    df = xl.parse(sheet)
    df['สถานะ'] = ''
    df['มหาวิทยาลัย'] = sheet
    df["คุณวุฒิ" ] =""
    df['คามเชี่ยวชาญ'] = ''
    df['สถานที่ติดต่อ'] = ''
    df = df[cols]
    
    dfs = pd.concat([dfs, df])
dfs.columns = new_cols
dfs                                                                                                         
```

```python
dfs.to_csv('tu_list.csv', encoding='utf-8-sig')
```
